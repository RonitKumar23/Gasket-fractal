```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=[0,0]
b=[2,0]
c=[1,2]
```

```python
def tri(a,b,c):
    plt.plot((a[0],b[0]),(a[1],b[1]))
    plt.plot((b[0],c[0]),(b[1],c[1]))
    plt.plot((c[0],a[0]),(c[1],a[1]))
```

```python
def midpoint(x,y):
    return (x[0]+y[0])/2 , (x[1]+y[1])/2
```

```python
def Gasket(a,b,c,n):
    
    tri(a,b,c)
    
    d=midpoint(a,b)
    e=midpoint(b,c)
    f=midpoint(c,a)
    
    n=n+1
    if n<6:
        tri(a,d,f)
        Gasket(a,d,f,n)
        
        tri(d,b,e)
        Gasket(d,b,e,n)
        
        tri(f,e,c)
        Gasket(f,e,c,n)
        
Gasket(a,b,c,0)
```
